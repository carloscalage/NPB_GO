```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/carloscalage/NPB_GO/main/results/log.csv')
df
```

```python
kernels = ['EP', 'IS']
classes = ['S', 'W', 'A']

for kernel in kernels:
  for c in classes:
    df_kernel = df[df['kernel'] == kernel]
    df_kernel['time'] = df_kernel['time'].apply(lambda x: x/1000000)
    df_c = df_kernel[(df_kernel['class'] == c) & (df_kernel['language'] == 'C')]['time'].to_numpy()
    df_go = df_kernel[(df_kernel['class'] == c) & (df_kernel['language'] == 'GO')]['time'].to_numpy()
    plt.figure(figsize=(15, 15))
    plt.rc('font', size=24)
    plt.ylabel('Time (ms)')
    plt.boxplot([df_c, df_go], labels=['C', 'GO'], widths=0.6)
    plt.title(f'Kernel {kernel} Class {c}')
    plt.savefig(f'results/{kernel}_{c}.png')
    plt.close()
```
